# Обучение с учителем: качество модели

В данном проекте будут исследованы данные интернет-магазина «В один клик», активность покупателей которого начала снижаться. Руководство определило путь развития - удержание активности постоянных клиентов с помощью персонализированных предложений. 

Цель исследования - разработать модель машинного обучения, которая будет предсказывать вероятность снижения покупательской активности. На основании предсказаний модели компания сможет предлагать персонализированные маркетинговые решения для удержания того или иного клиента. Продемонстрируем возможности полученной модели на дополнительном этапе исследования, отделив небольшой сегмент покупателей, для которых сделаем предсказания, и проанализируем, как увеличить покупательскую активность клиентов, предрасположенных к ее снижению.

В ходе исследования 
- ознакомимся с данными; 
- выполним предобработку данных: удалим явные/неявные дубликаты, проверим наличие пропусков, поработаем над типами данных;
- для исследования отберем клиентов с покупательской активностью не менее трёх месяцев,
- исследуем данные: проведем статистический анализ всех признаков, визуализируем данные с помощью графиков, избавимся от аномалий и выбросов;
- проведем корреляционный анализ данных: выясним, как входные признаки связаны с целевым - рассчитаем коэффициенты корреляции, проверим признаки на мультиколлинеарность;
- создадим модель машинного обучения: целевой признак - покупательская активность, будем использовать один общий пайплайн для перебора моделей и инструмент подбора гиперпараметров, который вернёт нам лучшую модель, подходящую для оценки качества метрику подберем в соответствие с особенностями предоставленных данных;
- проведем оценку важности признаков для полученной модели и построим график важности с помощью метода SHAP;
- выполним сегментацию покупателей и преведем пример использования полученной модели для разработки маркетинговых предложений;
- сформулируем общие выводы о проделанной работе, разработанной модели и выделенном сегменте покупателей.

### Изменение типов данных

```python
def to_float(row): #создадим функцию преобразования строк
    try:
        if ',' in row:
            return float(row.replace(',', '.'))
        else:
            return float(row)
    except:
        return 'ошибка'
```

```python
profit['прибыль'] = profit['прибыль'].apply(to_float) #используем написанную функцию для преобразования
first_look(profit)
```

## Исследовательский анализ данных

На этом этапе исследования подробно рассмотрим данные всех датафреймов, проверим их на наличие выбросов, изучим особенности их распределений. Проведем стастический анализ. <br>
Перед тем как начать, сразу отберем клиентов с покупательской активностью не менее трёх месяцев в соответствие с ТЗ - так как текущий месяц еще не окончен, в наборе данных будем оставлять клиентов, которые совершали покупки в каждом из двух предыдущих месяцев.

Для этого сначала видоизменим датафреймы market_money и market_time для удобства фильтрации и последующего объединения таблиц.

```python
#создаем сводную таблицу
market_money = market_money.pivot_table(index='id', columns='период', values='выручка').reset_index(drop = False)
market_money.columns = ['id', 'предыдущий_месяц', 'препредыдущий_месяц', 'текущий_месяц']
market_money.head(10) #выводим 10 строк
```

```python
#создаем сводную таблицу
market_time = market_time.pivot_table(index='id', columns='период', values='минут').reset_index(drop = False)
market_time.columns = ['id', 'предыдущий_месяц', 'текущий_месяц']
market_time.head(10) #выводим 10 строк
```

Теперь можно приступать к фильрации: найдем клиентов, которые не совершали покупок в предыдущем или препредыдущем месяцах.

```python
market_money.query('предыдущий_месяц == 0 or препредыдущий_месяц == 0')
```

Было обнаружено 3 клиента, покупательская активность которых не соответсвует требованиям закачика. Удалим их из всех датафреймов.

```python
def filter_df(df): #создаем функцию фильтрации
    df = df.query('~(id == 215348 or id == 215357 or id == 215359)')
    return df
```

```python
#применяем функицию для фильрации всех датафреймов, смотрим размерность
market_clients = filter_df(market_clients)
print(market_clients.shape)

market_money = filter_df(market_money)
print(market_money.shape)

market_time = filter_df(market_time)
print(market_time.shape)

profit = filter_df(profit)
print(profit.shape)
```

Олично теперь во всех датафреймах количество строк соответствует количеству исследуемых покупателей, следовательно, фильтрацию провели корректно. Можно приступать к исследовательскому анализу.

## Использование пайплайнов

Для создания оптимальной модели машинного обучения будем использовать один общий пайплайн для всех моделей. Выбирать будем из KNeighborsClassifier, DecisionTreeClassifier, LogisticRegression и  SVC, для каждой подберем один или несколько гиперапараметров. <br>
Для кодирования категориальных признаков будем использовать два кодировщика - OneHotEncoder и OrdinalEncoder, для масштабирования количественных попробуем StandardScaler и MinMaxScaler. <br>
Для оценки результатов работы модели стоит выбирать между F1 и ROC-AUC метриками. Обе они могут использоваться для несбалансированных классов, однако f1-мера в данном случае будет более подходящей, так как здесь нам важен баланс между точностью и полнотой. Нам нужно как можно больше покупателей правильно определить, как тех кто снизит покупательскую активность, ведь если мы пропустим клиента, у которого снизится активность, то это приведет к потере потенциальной прибыли. И нам также важно, чтобы те, кто были предсказаны, как клиенты у которых снизится покупательская активность, были именно таковыми. Иначе на таких клиентов будут впустую потрачены ресурсы на маркетинг, хотя они в этом не нуждались, т.е. компания понесет ненужные издержки.

```python
#делим данные на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    market_full.drop(['покупательская_активность'], axis=1),
    market_full['покупательская_активность'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = market_full['покупательская_активность'])

#кодируем целевую переменную
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)

# создаём списки с названиями признаков
ohe_columns = ['разрешить_сообщать', 'популярная_категория']
ord_columns = ['тип_сервиса']
num_columns = ['маркет_актив_6_мес', 'маркет_актив_тек_мес', 'длительность', 'акционные_покупки',
               'средний_просмотр_категорий_за_визит', 'неоплаченные_продукты_штук_квартал',
               'ошибка_сервиса', 'страниц_за_визит', 'выручка_предыдущий_месяц', 'выручка_препредыдущий_месяц',
               'выручка_текущий_месяц', 'время_предыдущий_месяц', 'время_текущий_месяц']

# создаём пайплайн для подготовки признаков из списка ohe_columns: OHE-кодирование
# OHE
ohe_pipe = Pipeline(
    [
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ]
    )

# создаём пайплайн для подготовки признаков из списка ord_columns: Ordinal-кодирование
# OE
ord_pipe = Pipeline(
    [
     ('ord',  OrdinalEncoder(
                categories=[
                    ['стандарт', 'премиум']
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
    ]
)

# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(5, 12),
        'models__max_features': range(5, 15),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(2, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # словарь для модели SVC()
    {
        'models': [SVC(kernel = 'linear', random_state=RANDOM_STATE, probability=True)],
        'models__C': range(2, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  

    },
]

randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='f1',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)


print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print(randomized_search.best_params_)
print ('Метрика лучшей модели на тренировочной выборке:', randomized_search.best_score_)

# проверяем работу модели на тестовой выборке
# рассчитываем прогноз на тестовых данных
y_test_pred = randomized_search.predict(X_test)
print(f'F1-мера на тестовой выборке: {f1_score(y_test, y_test_pred)}')
```

Итак, лучшей моделью из предложенных оказалась SVC с линейным ядром и "силой" модели С=6, для масштабирования использован StandardScaler. На тестовой выборке модель показала даже лучшие результаты, чем на тренировочной, что может говорить нам об отсутствии факта переобучения.

## Анализ важности признаков

```python
#трансофрмируем признаки так же, как и при обучении лучшей модели
X_train_transformed = pipe_final.named_steps['preprocessor'].fit_transform(X_train)
explainer = shap.LinearExplainer(randomized_search.best_estimator_.named_steps['models'], X_train_transformed)

#задаем имена признаков
feature_names = randomized_search.best_estimator_.named_steps['preprocessor'].get_feature_names_out()
explainer.feature_names = feature_names

#получаем значения shap
shap_values = explainer(X_train_transformed)

#строим график
shap.plots.beeswarm(shap_values, max_display = 20) 
```

Судя по полученным графикам, наибольшей значимостью, при предсказании снижения покупательской активности, обладает категория товара, которая наболее интересна клиенту: вперед выбились такие категории товаров, как "мелкая бытовая техника и электроника", "техника для красоты и здоровья" и "товары для детей". <br>
Разберем на примере бытовой техники. В нашем случае более высокое значение признака Мелкая бытовая техника и электроника склоняет модель к классу 0 (прежний уровень). <br>
То есть клиенты, покупающие/интересующиеся мелкой бытовой техникой, электроникой, техникой для красоты и товарами для детей с меньшей вероятностью снизят свою покупательскую активность, нежели покупатели, которые чаще покупают косметику или посуду в интернет магазине «В один клик».  <br>
Следующими важными показателями  являются количества просмотренных страниц и категорий товаров за визит (средние SHAP‑значения 0.08). Чем больше значение признака, тем вероятнее модель решит, что клиент не потеряет интерес к магазину в ближайшее время. Эта же зависимость прослеживается и со временем, проведенном на сайте в предыдущем и текущем месяцах (их средние SHAP‑значения близки - 0.08 и 0.06) : чем больше человек просматривает категорий, пролистывает страниц, чем больше задерживается на сайте, тем вероятнее прогноз модели для него будет нулевым - прежний уровень покупательской активности. <br>
Интереcный результат дал показатель "акционные покупки", будто бы отделился сегмент покупателей с высокой долей акционных товаров (как было замечено и ранее в исследовании), этому сегменту соответсвуют большие средние SHAP‑значения, то есть это является ярким индикатором для прогноза, что клиент снизит покупательскую активность. Это легко можно представить в реальной жизни: клиент ищет определенный товар по разным сайтам, как только он продается по акции на какой-то площадке, клиент его покупает, при этом не обретая или же совсем малый интерес к самому магазину, то есть ждать от него активности в следующий месяц не стоит. <br>
Интересно также, что для модели клиенты с типом сервиса "премиум" более склонны к снижению покупательской активности. Здесь нет очевидного объяснения, может по данной группе покупателей стоит провести отдельное всестороннее исследование. Может качество сервиса не соответствует запросу, может на сайте представлено не так много товаров, интересующих людей с таким типом сервиса, поэтому после одной-двух покупок у них снижается интерес. Однако для нашей модели этот признак относится скорее к маловажным - среднее SHAP‑значение 0.04. <br>
Также среди менее важных признаков для модели оказались выручка (за любой месяц), среднемесячное значение маркетинговых коммуникаций, ошибки сервиса. Практически неважно для модели в прогнозе длительность пребывания на сайте (с момента регистрации) и маркетинговая активность этого месяца.
